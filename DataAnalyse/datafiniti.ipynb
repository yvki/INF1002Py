```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# bargraph 
def map_sentiment(rating):
    if rating < 2.5:
        return 'Negative' 
    elif rating == 2.5:
        return 'Neutral'
    else:
        return 'Positive'

df = pd.read_csv('../CSVInUse/datafiniti.csv', index_col=0)
# drop duplicated index column
df = df.iloc[:, 1:]
df['sentiment'] = [map_sentiment(x) for x in df['reviews.rating']]
sns.countplot(x='sentiment', data=df)
plt.show()
```

```python
# scatterplot
import matplotlib.pyplot as plt
df.plot(x='latitude', y='longitude', style='x')  
plt.title('Latitude and Longitude')  
plt.xlabel('latitude')  
plt.ylabel('longitude')  
plt.show() 
```

```python
# lineplot
def map_sentiment(rating):
    if rating < 2.5:
        return 'Negative' 
    elif rating == 2.5:
        return 'Neutral'
    else:
        return 'Positive'

df = pd.read_csv('datafiniti.csv', index_col=0)
# drop duplicated index column
df = df.iloc[:, 1:]
df['sentiment'] = [map_sentiment(x) for x in df['reviews.rating']]
df['year'] = pd.DatetimeIndex(df['reviews.date']).year
sns.lineplot(data=df, x='year', y='sentiment')
plt.show()
```
