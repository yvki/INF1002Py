```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import time
from time import sleep
from random import randint
import matplotlib.pylab as plt
```

```python
page = np.arange(5, 31, 5)  

name = []
title = []
review = []
rating = []

for p in page:
    #url = 'https://www.tripadvisor.com.sg/Hotel_Review-g294265-d4456040-Reviews-or' + str(p) + '-Holiday_Inn_Express_Singapore_Orchard_Road-Singapore.html#REVIEWS'
    #url = 'https://www.tripadvisor.com.sg/Hotel_Review-g35805-d250023-Reviews-or' + str(p) + '-River_Hotel_Chicago-Chicago_Illinois.html#REVIEWS'
    #url = 'https://www.tripadvisor.com.sg/Hotel_Review-g32655-d8552209-Reviews-or' + str(p) + '-Hotel_Indigo_Los_Angeles_Downtown_an_IHG_Hotel-Los_Angeles_California.html#REVIEWS'
    url = 'https://www.tripadvisor.com.sg/Hotel_Review-g298307-d1168669-Reviews-or' + str(p) + '-Pantai_Inn-Kota_Kinabalu_Kota_Kinabalu_District_West_Coast_Division_Sabah.html#REVIEWS'

    req = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = req.content
    time.sleep(2)    
    soup = BeautifulSoup(html, 'html.parser')

    # crawl customer name
    for a in soup.find_all('a', class_='ui_header_link uyyBf'):
        name.append(a.get_text(strip=True))
        
    # crawl review title
    for a in soup.find_all('a', class_='Qwuub'):
        title.append(a.get_text(strip=True))

    # crawl review text
    for a in soup.find_all('q', class_='QewHA H4 _a'):
        review.append(a.get_text())

    # crawl review rating only from bubble 50 not bubble ui
    for a in soup.find_all('div', class_='IkECb f O'):
        rating.append(a.span['class'])

final_rating = []
for i in rating:
    bubble_rating = str(i[1])
    final_rating.append(int(bubble_rating[-2]))
# check array length    
print(len(name), len(final_rating), len(title), len(review))
```

```python
# run only for holiday inn, river hotel, hotel indigo
df = pd.DataFrame({'Customer_name': name,'Rating': final_rating,'Review_Title': title,'Review': review})
df.to_csv('../CSVInUse/PantaiInnReview.csv') #change csv file name
df_new = pd.read_csv('../CSVInUse/PantaiInnReview.csv') #change csv file name 
df_new
```
